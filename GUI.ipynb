```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tkinter import Tk, Label, Button, messagebox,Frame,filedialog
from tkinter import Tk, Label
from PIL import Image, ImageTk
import tkinter.font as tkfont
```

```python

cnn=load_model("lcd.h5")
```

```python
# Create the main window
window = Tk()
window.title("Lung Cancer Detection")
font_style = tkfont.Font(family="Helvetica", size=12, weight="bold")
# Function to load and classify the image
def load_image():
    file_path = filedialog.askopenfilename()
    if file_path:
        try:
            # Open and display the loaded image
            img = Image.open(file_path)
            img.thumbnail((300, 300))  # Resize the image to fit the window
            img_tk = ImageTk.PhotoImage(img)
            image_label.configure(image=img_tk)
            image_label.image = img_tk
        except Exception as e:
            messagebox.showerror("Error", str(e))
    else:
        messagebox.showwarning("Warning", "No image selected.")

# Function to classify the loaded image
def classify_image():
    image_data = image_label.image
    if image_data:
        try:
            # Convert PhotoImage to PIL Image
            img_pil = ImageTk.getimage(image_data)
            
            # Resize and preprocess the image
            img_pil = img_pil.resize((64, 64))  # Resize the image for model input
            img_pil = img_pil.convert("RGB")
            img = image.img_to_array(img_pil)
            img = img / 255.0  # Normalize the image data
            img = np.expand_dims(img, axis=0)
            
            # Make a prediction using the CNN model
            result = cnn.predict(img)
            prediction = "No Cancer Detected" if result[0][0] > 0.5 else "Cancer Detected"
            
            # Display the prediction
            result_label.config(text=prediction)
        except Exception as e:
            messagebox.showerror("Error", str(e))
    else:
        messagebox.showwarning("Warning", "No image loaded.")

# Function to close the application
def close_app():
    if messagebox.askokcancel("Quit", "Do you want to quit?"):
        window.destroy()

# Load the background image
bg_image = Image.open("2657595.jpg")
window_width = window.winfo_screenwidth()
window_height = window.winfo_screenheight()
bg_image = bg_image.resize((window_width, window_height))
bg_photo = ImageTk.PhotoImage(bg_image)
bg_label = Label(window, image=bg_photo)
bg_label.place(x=0, y=0)

# Create a label for the loaded image
image_label = Label(window,font=font_style,fg="white", bg="red")
image_label.pack(pady=10)

# Button to load an image
load_button = Button(window, text="Load Image", command=load_image,font=font_style,fg="white", bg="red")
load_button.pack(pady=10)

# Button to classify the loaded image
classify_button = Button(window, text="Predict", command=classify_image,font=font_style,fg="white", bg="red",)
classify_button.pack(pady=10)

# Label to display the prediction result
result_label = Label(window, text="",font=font_style)
result_label.pack(pady=10)

# Button to close the application
close_button = Button(window, text="Close", command=close_app,font=font_style,fg="white", bg="red")
close_button.pack(pady=10)
# Label for the lung cancer prediction description
description_label = Label(window, text="Lung Cancer Prediction",font=font_style)
description_label.config(font=("Roboto", 16))
description_label.pack(pady=20)

# Label for the description text
text = "This application uses a Convolutional Neural Network (CNN) model to predict\n" \
       "whether a given image of a lung contains cancerous cells or not.\n\n" \
       "Instructions:\n" \
       "1. Click the 'Load Image' button to select an image.\n" \
       "2. Once the image is loaded, click the 'Predict' button to classify it.\n" \
       "3. The prediction result will be displayed below.\n\n" \
       
description_text_label = Label(window, text=text, justify="left",font=font_style,fg="black",bg=window.cget('bg'), highlightthickness=0, bd=0)
#description_text_label.configure(background='transparent')
description_text_label.configure(background=window.cget('bg')) 
description_text_label.pack(pady=10)


# Run the GUI
window.mainloop()
```
